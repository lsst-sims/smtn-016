```python
import rubin_sim.maf as maf
import matplotlib.pylab as plt
%matplotlib inline
from rubin_sim.data import get_baseline
import os
```

```python
baseline_file = 'baseline_v2.1_10yrs.db'
runName = os.path.split(baseline_file)[-1].replace('.db', '')
out_dir='temp'
```

```python
bundleList = []
summary_list = [maf.metrics.MedianMetric()]
slicer = maf.slicers.HealpixSlicer()
for filtername in 'ugrizy':
    sql = 'filter = "%s"' % filtername
    metric = maf.metrics.SurfaceBrightLimitMetric(metricName='sb limit, %s' % filtername)
    bundleList.append(maf.MetricBundle(metric, slicer, constraint=sql, summaryMetrics=summary_list, runName=runName))
```

```python
bg = maf.MetricBundleGroup(bundleList, baseline_file, out_dir, None)
bg.runAll()
bg.plotAll(closefigs=False)
```

```python
for b in bundleList:
    print(b.constraint, b.summaryValues)
```
